Creamos los DataFrame para trabajar de los archivos seleccionados.

```python
df_sales = pd.read_csv(ruta_sales, parse_dates=["SalesDate"] )
df_employees = pd.read_csv(ruta_employees, parse_dates=["BirthDate", "HireDate"])
df_products = pd.read_csv(ruta_products)
```

Analizando los dataframes cargados para entender su estructura y contenido.

```python
print("Información de df_sales:")
display(df_sales.info())
print("\nPrimeras filas de df_sales:")
display(df_sales.head())

print("\nInformación de df_employees:")
display(df_employees.info())
print("\nPrimeras filas de df_employees:")
display(df_employees.head())

print("\nInformación de df_products:")
display(df_products.info())
print("\nPrimeras filas de df_products:")
display(df_products.head())
```

Calculando el valor real de la venta (TotalPriceCalculated) y almacenándolo en una nueva columna.

```python
# Merge df_sales with df_products to get product price
df_sales = pd.merge(df_sales, df_products[['ProductID', 'Price']], on='ProductID', how='left')

# Calculate the actual total price
df_sales['TotalPriceCalculated'] = df_sales['Quantity'] * df_sales['Price']

print("DataFrame df_sales con TotalPriceCalculated:")
display(df_sales.head())
```

Detectando outliers en la columna de ventas totales (`TotalPriceCalculated`) utilizando el criterio del rango intercuartílico (IQR) y marcándolos con una nueva columna (`IsOutlier`).

```python
Q1 = df_sales['TotalPriceCalculated'].quantile(0.25)
Q3 = df_sales['TotalPriceCalculated'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_sales['IsOutlier'] = ((df_sales['TotalPriceCalculated'] < lower_bound) | (df_sales['TotalPriceCalculated'] > upper_bound)).astype(int)

num_outliers = df_sales['IsOutlier'].sum()
print(f"Se detectaron {num_outliers} outliers en la columna TotalPriceCalculated.")
display(df_sales.head())
```

Clasificando las ventas como "Entre semana" o "Fin de semana" a partir de `SalesDate` y comparando el total de ventas entre ambos grupos.

```python
df_sales['DayOfWeek'] = df_sales['SalesDate'].dt.dayofweek
df_sales['Weekday_Weekend'] = df_sales['DayOfWeek'].apply(lambda x: 'Fin de semana' if x >= 5 else 'Entre semana')

sales_by_day_type = df_sales.groupby('Weekday_Weekend')['TotalPriceCalculated'].sum().sort_values(ascending=False)

print("Total de ventas por día de la semana vs. Fin de semana:")
display(sales_by_day_type)

most_sales_day_type = sales_by_day_type.index[0]
print(f"\nLa empresa vende más durante: {most_sales_day_type}.")
```

Calculando la edad del empleado al momento de su contratación y los años de experiencia al momento de cada venta.

```python
df_merged = pd.merge(df_sales, df_employees[['EmployeeID', 'BirthDate', 'HireDate']], left_on='SalesPersonID', right_on='EmployeeID', how='left')

df_merged['AgeAtHire'] = (df_merged['HireDate'] - df_merged['BirthDate']).dt.days / 365.25
df_merged['ExperienceAtSale'] = (df_merged['SalesDate'] - df_merged['HireDate']).dt.days / 365.25

display(df_merged[['SalesPersonID', 'SalesDate', 'BirthDate', 'HireDate', 'AgeAtHire', 'ExperienceAtSale']].head())
```

Preparando un único dataset para modelado combinando información relevante y aplicando transformaciones.

```python
# Combinar df_merged con df_products para incluir información del producto
df_final = pd.merge(df_merged, df_products[['ProductID', 'ProductName', 'CategoryID', 'Class', 'Resistant', 'IsAllergic', 'VitalityDays']], on='ProductID', how='left')

# Extract SalesHour from SalesDate
df_final['SalesHour'] = df_final['SalesDate'].dt.hour

# Seleccionar columnas relevantes para el modelado
df_final = df_final[['SalesPersonID', 'CustomerID', 'ProductID', 'Quantity', 'Discount',
                     'TotalPriceCalculated', 'SalesDate', 'TransactionNumber',
                     'DayOfWeek', 'Weekday_Weekend', 'SalesHour', 'IsOutlier',
                     'AgeAtHire', 'ExperienceAtSale', 'ProductName', 'CategoryID',
                     'Class', 'Resistant', 'IsAllergic', 'VitalityDays']]

# Eliminar la columna EmployeeID duplicada después del merge
df_final = df_final.drop('EmployeeID', axis=1, errors='ignore')


# Transformaciones para variables categóricas (One-Hot Encoding)
categorical_cols = ['Weekday_Weekend', 'Class', 'Resistant', 'IsAllergic']
df_final = pd.get_dummies(df_final, columns=categorical_cols, drop_first=True)

# Transformaciones para variables numéricas (Ejemplo: escalado - opcional, dependiendo del modelo)
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# numerical_cols = ['Quantity', 'Discount', 'SalesHour', 'AgeAtHire', 'ExperienceAtSale', 'VitalityDays']
# df_final[numerical_cols] = scaler.fit_transform(df_final[numerical_cols])

print("Dataset final preparado para modelado:")
display(df_final.head())
display(df_final.info())
```
